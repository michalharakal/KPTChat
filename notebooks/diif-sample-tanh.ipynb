```kotlin
%use kandy
%use dataframe
```

```kotlin
import de.jugda.knanogpt.core.tensor.Shape
import de.jugda.knanogpt.core.tensor.Tensor
import org.skainet.diff.D
```

```kotlin
import kotlin.random.Random

val random = Random(1337)

// Define the range
val min = -2 * PI
val max = 2 * PI

// Generate a list of 100 random values within the range -2*PI to 2*PI
val randomValues = List(100) { min + random.nextDouble() * (max - min) }

//random.seed(19)
val x_ten = Tensor(Shape(randomValues.size), randomValues.toDoubleArray())
val y_cal = randomValues.map { sin(it.toDouble()).toDouble() }
val x = D(x_ten, Tensor(x_ten.shape, x_ten.elements))
```

```kotlin
x
```

```kotlin
import org.skainet.diff.grad
import org.skainet.diff.sins

val y = grad { sins(x) }
y
```

```kotlin
val x_values by columnOf(*x.x.elements.toTypedArray())
val y_values by columnOf(*y.d.elements.toTypedArray())
val y_values_cal by columnOf(*y_cal.toTypedArray())


val df = dataFrameOf(x_values, y_values, y_values_cal)
df
```

```kotlin
df.plot {
    line {
        x(x_values)
        y(y_values) 
    }
    line {
        x(x_values)
        y(y_cal)
    }
}
```
